```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import codecs
```

```python
# change the path down here for a different tournament

file=codecs.open("htmlFiles/cmu_cards.html", 'r')
fileStr = file.read()
```

```python
soup = bs(fileStr, 'html.parser')
# print(soup.prettify())
body = soup.body
```

```python
children = body.findChildren("table") # , recursive=False)
childrenP = body.findChildren("p") # , recursive=False)
```

```python
teams = []

for p in childrenP:
    text = p.text

    #print(text[:6])
    if text[:6] == 'Team: ':
        teams.append(text[6:])

```

```python
bigKidArray = []
```

```python
teamCounter = 0

for teamTable in children:

    innerCounter = 0
    currArray = []

    # ADDED TO ARRAY IN THIS ORDER
    currTeamName = "none"
    currWins = 0
    currGovs = 0
    currDebater1 = "none"
    currDebater1Status = "none"
    currDebater1Speaks = -1
    currDebater1Ranks = -1
    currDebater2 = "none"
    currdebater2Status = "none"
    currDebater2Speaks = -1
    currDebater2Ranks = -1

    for tr in teamTable.findChildren("tr"):
        # print("Inner counter:", innerCounter)
        row = innerCounter

        #print(len(tr.findChildren("td")))
        #if row == 0
        loTD = tr.findChildren("td")

        if innerCounter == 0:
            debater1Block = loTD[5].text
            name1 = debater1Block[:-4]
            status1 = debater1Block[-3:]
            status1 = status1.replace('(','').replace(')','')
            
            debater2Block = loTD[6].text
            name2 = debater2Block[:-4]
            status2 = debater2Block[-3:]
            status2 = status2.replace('(','').replace(')','')
            
            currDebater1 = name1
            currDebater1Status = status1
            currDebater2 = name2
            currDebater2Status = status2
        
        if innerCounter > 0 and innerCounter <  6:
            side = loTD[1].text
            result = loTD[2].text


            if side == 'G':
                currGovs +=1
            if result == 'W':
                currWins +=1

        if innerCounter == 6:
            debater1EndLine = loTD[1].text
            debater2EndLine = loTD[2].text

            currDebater1Speaks = float(debater1EndLine.split(",")[0].replace('(','').replace(')','').strip())
            currDebater1Ranks = float(debater1EndLine.split(",")[1].replace('(','').replace(')','').strip())
            currDebater2Speaks = float(debater2EndLine.split(",")[0].replace('(','').replace(')','').strip())
            currDebater2Ranks = float(debater2EndLine.split(",")[1].replace('(','').replace(')','').strip())


        #for td in tr.findChildren("td"):
        #    print("Cell:", td.text)

        if innerCounter == 6:
            teamSummaryArray = []
            teamSummaryArray.append(teams[teamCounter])
            teamSummaryArray.append(currWins)
            teamSummaryArray.append(currGovs)
            teamSummaryArray.append(currDebater1)
            teamSummaryArray.append(currDebater1Status)
            teamSummaryArray.append(currDebater1Speaks)
            teamSummaryArray.append(currDebater1Ranks)
            teamSummaryArray.append(currDebater2)
            teamSummaryArray.append(currDebater2Status)
            teamSummaryArray.append(currDebater2Speaks)
            teamSummaryArray.append(currDebater2Ranks)

            bigKidArray.append(teamSummaryArray)

            innerCounter = 0

            # reset vals
            currTeamName = "none"
            currWins = 0
            currGovs = 0
            currDebater1 = "none"
            currDebater1Status = "none"
            currDebater1Speaks = -1
            currDebater1Ranks = -1
            currDebater2 = "none"
            currdebater2Status = "none"
            currDebater2Speaks = -1
            currDebater2Ranks = -1

            break
        else:
            innerCounter += 1
    
    teamCounter += 1
```

```python
df = pd.DataFrame(bigKidArray, columns = ['team_name', 'wins', 'govs', 'debater_1_name', 'debater_1_status', 'debater_1_speaks', 'debater_1_ranks', 'debater_2_name', 'debater_2_status', 'debater_2_speaks', 'debater_2_ranks']) 
df.head(34)
```

```python
df.to_csv('export_csv.csv')
```
